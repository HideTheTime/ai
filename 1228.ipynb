```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()


print(pd)
```

```python
data2  = pd.read_csv('generated_data.csv')
print(data2)
```

```python
x = data2.loc[:,'x']
y = data2.loc[:,'y']
x = np.array(x).reshape(-1,1)
y = np.array(y).reshape(-1,1)
print(y,x)
from matplotlib import pyplot as plt
plt.figure(figsize=(20,20))
plt.scatter(x,y)
plt.show()
```

```python

print(x.shape,type(x))
print(y,x)
```

```python
lr_model.fit(x,y)
```

```python
y_predict = lr_model.predict(x)
print(y_predict)
```

```python
a = lr_model.coef_
b = lr_model.intercept_
print(a,b)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
# 上面的是均方差和R方值
# print(mean_squared_error)
print(y_predict)
MSE = mean_squared_error(y,y_predict)
R2 = r2_score(y,y_predict)
print(MSE,R2)
```

```python
plt.figure()
plt.plot(y,y_predict)
plt.show(  )
```

```python
data_house = pd.read_csv('usa_housing_price.csv')
# print(data_house)
size = data_house.loc[:,'size']
price = data_house.loc[:,'Price']
area_income = data_house.loc[:,'Avg. Area Income']
# print(area_income)
ax = plt.figure(figsize=(20,20)).gca(projection='3d')
ax.legend()
ax.scatter(price,size,area_income)
# plt.scatter()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
LR1 = LinearRegression()

print(size.shape)

x = np.array(size).reshape(-1,1)
y = np.array(price).reshape(-1,1)
print(x,y)
```

```python
fit1 = LR1.fit(x,y)
print(fit1)
# 训练完成
```

```python
y_predict = LR1.predict(x)
print(y_predict)
```

```python
# mse和r方
from sklearn.metrics import mean_squared_error,r2_score
y_mse = mean_squared_error(y,y_predict)
y_r2 = r2_score(y,y_predict)
print(y_mse,y_r2)
# 0.12R2表示结果并不好
```

```python
plt.figure()
plt.scatter(x,y)
plt.plot(x,y_predict,'r')
plt.show()
```
